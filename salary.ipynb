# *1. Import required libraries*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# *2. Load dataset from CSV file*

```python
dataset = pd.read_csv("Salary_Data.csv")

# Separate features (X) and target (y)
X = dataset.drop('Salary', axis=1).values
y = dataset['Salary'].values
```

```python
print(X)
```

```python
print(y)
```

# *3. Split the dataset into Training Set and Test Set*

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2)
```

# *4. Train the Linear Regression Model*

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# **Save Model**

```python
import joblib

joblib.dump(regressor, "model.pkl")
print("Model Saved Successfully")
```

# *5. Predict the Test Set Results*

```python
y_pred = regressor.predict(X_test)

print("Actual Values:")
print(y_test)

print("\nPredicted Values:")
print(y_pred)
```

# *6. Plot Training Data, Regression Line, Add title and label axis*

```python
plt.scatter(X_train, y_train, color='red', label='Training Data')
plt.plot(X_train, regressor.predict(X_train),color='blue',label='Regression Line')

plt.title('Salary Prediction using Linear Regression')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()

plt.show()
```
